```python
#Q3. Singular Value Decomposition (SVD):
#Perform Singular Value Decomposition on the matrix A obtained in Question 2. Separate and print matrices U, Î£, and ñ‘‰ . Verify that A equals the product of U, Î£, and ñ‘‰ . Additionally, find the rank 2 and rank 3 approximations of matrix A.
```

```python
#PERFORMING SVD ON MATRIX A
np.random.seed(23)
D= np.random.randint(0,10,(5,5))
sp.Matrix(D)
```

```python
U,S,VT=np.linalg.svd(D)
sp.Matrix(np.round(U))
```

```python

#creating sigma
sigma = np.zeros((D.shape[0], D.shape[1]))
sp.Matrix(sigma)
```

```python
sigma[:D.shape[1], :D.shape[0]] = np.diag(S)
sp.Matrix(np.round(sigma))
```

```python
sp.Matrix(np.round(VT))
```

```python
sp.Matrix(np.round(U@sigma@VT))
```

```python
#FINDING RANK 2 & RANK 3 APPROXIMATIONS
```

```python
Uk = U[:,:2]
# Choose all rows and first 2 columns
sigmak = sigma[:2,:2] # Choose first 2 rows and 2 columns
VTk = VT[:2,:] # Choose first 2 rows and all columns
Ak = Uk@sigmak@VTk
sp.Matrix(Ak)
```

```python
np.linalg.norm(D-Ak)
```

```python
Uk1 = U[:,:3]
# Choose all rows and first 3 columns
sigmak1 = sigma[:3,:3] # Choose first 3 rows and 3 columns
VTk1 = VT[:3,:] # Choose first 3 rows and all columns
Ak1 = Uk1@sigmak1@VTk1
sp.Matrix(Ak1)
```

```python
np.linalg.norm(D-Ak1)
```

```python
#Q4. Polynomial Plotting with Annotations:
#Randomly select the coefficients of a 5th degree polynomial. Set the random seed as the last two digits of your roll number. Plot the polynomial for -5 â‰¤ x â‰¤ 5. Annotate the plot to identify the maxima and minima of the polynomial
```

```python
import matplotlib.pyplot as plt
p = np.pi
```

```python
np.random.seed(23)
C = np.random.randn(6)
C
```

```python

X = np.arange(-5, 5, 0.1)
Y=C[0] + C[1]*X + C[2]*X**2 + C[3]*X**3 + C[4]*X**4 + C[5]*X**5
```

```python


# Find the minima and maxima points
minindex = np.argmin(Y)
maxindex = np.argmax(Y)
minpt = (X[minindex], Y[minindex])
maxipt = (X[maxindex], Y[maxindex])
```

```python
# Plot the polynomial function
plt.plot(X, Y, color="blue")

plt.annotate("minima", xy=minpt, xytext=(2.5, -1000), arrowprops=dict(facecolor="red", shrink=1))
plt.annotate("maxima", xy=maxipt, xytext=(-3.5,3000), arrowprops=dict(facecolor="black", shrink=1))

plt.show()
```
